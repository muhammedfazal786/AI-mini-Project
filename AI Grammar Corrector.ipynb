# AI Grammar Corrector

```python
from openai import OpenAI
```

```python
client = OpenAI(base_url="http://localhost:11434/v1",api_key="ollama")
```

```python
while True:
    text = input("Enter sentence to correct (type 'exit' to stop): ")

    if text.lower() == "exit":
        break

    response = client.chat.completions.create(
        model="llama3:8b",
        messages=[
            {
                "role": "system",
                "content": "You are an AI grammar corrector. Correct the grammar of the given sentence and return the improved sentence."
            },
            {
                "role": "user",
                "content": text
            }
        ]
    )

    print("\nCorrected Sentence:\n")
    print(response.choices[0].message.content)
    print("\n" + "-"*50)
```
