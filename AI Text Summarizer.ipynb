```python
from openai import OpenAI
```

```python
client = OpenAI(base_url="http://localhost:11434/v1",api_key="ollama")
```

# AI Text Summarizer

```python
while True:
    text = input("Enter text to summarize (type 'exit' to stop): ")

    if text.lower() == "exit":
        break

    response = client.chat.completions.create(
        model="llama3:8b",
        messages=[
            {
                "role": "system",
                "content": "You are an AI text summarizer. Summarize the given text clearly and briefly."
            },
            {
                "role": "user",
                "content": f"Summarize this text:\n{text}"
            }
        ]
    )

    print("\nSummary:\n")
    print(response.choices[0].message.content)
    print("\n" + "-"*50)
```
