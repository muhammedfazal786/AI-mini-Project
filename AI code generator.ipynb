```python
from openai import OpenAI
```

```python
client = OpenAI(base_url="http://localhost:11434/v1",api_key="ollama")
```

# AI code generator

```python
while True:
    user = input("Enter your coding problem: ")

    if user.lower() == "exit":
        break

    response = client.chat.completions.create(
        model="llama3:8b",
        messages=[
            {
                "role": "system",
                "content": "You are an AI code generator. When the user asks a programming problem, directly generate the complete code solution with a short explanation."
            },
            {
                "role": "user",
                "content": user
            }
        ]
    )

    print("\nGenerated Code:\n")
    print(response.choices[0].message.content)
```
