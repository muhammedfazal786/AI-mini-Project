# AI Email Writer

```python
from openai import OpenAI
```

```python
client = OpenAI(base_url="http://localhost:11434/v1",api_key="ollama")
```

```python
while True:
    topic = input("Enter the email topic or request (type 'exit' to stop): ")

    if topic.lower() == "exit":
        break

    response = client.chat.completions.create(
        model="llama3:8b",
        messages=[
            {
                "role": "system",
                "content": "You are an AI email writing assistant. Write clear, professional, and polite emails."
            },
            {
                "role": "user",
                "content": f"Write a professional email about: {topic}"
            }
        ]
    )

    print("\nGenerated Email:\n")
    print(response.choices[0].message.content)
    print("\n" + "-"*50)
```
